```python
# Importing the pathlib and csv libraries
from pathlib import Path
import csv

# Setting the file path
csvpath = Path('Resources/budget_data.csv')

# Assigning a variable to hold columns as lists
date = []
profit_loss = []

# Opening the budget_data resource file 
with open(csvpath) as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    
    # Skipping headers
    column = next(csvreader, None)
    
    # Reading one data set at a time
    for column in csvreader:
        
        # Treating each column as separate list
        date.append(column[0])
        profit_loss.append(int(column[1]))
        
# Assigning Variables to enable computations
count_months = 1
sum_profit_loss = 0
profitloss_change = 0
summation_change = 0
average_change = 0
max_increase = 0
max_decrease = 0
max_inc_month = 1
max_dec_month = 1

    
#Introducing a check to count months based on the length of data points in profi/loss column
while count_months < len(profit_loss):
    
    #Calculating Total profit/loss
    sum_profit_loss += sum(profit_loss)
    
    # Calculating change in profits as current month minus previous month
    profitloss_change = profit_loss[count_months] - profit_loss[count_months-1]
    
    # Computing Average change as Summation of change in profits divided by count of profit/loss change
    summation_change += profitloss_change
    average_change = round(summation_change/(count_months),2)
    

    
    #Introducing conditions to determing greatest increase in profits
    if  profitloss_change > max_increase:
        max_increase = profitloss_change
        max_inc_month = count_months
    
    elif profitloss_change < max_decrease:
        max_decrease = profitloss_change
        max_dec_month = count_months
        
    # Increment counter
    count_months += 1

# Printing the analysis to the terminal
print(f"Financial Analysis:")
print("----------------------------")
print(f"Total Months: {count_months}")
print(f"Total: {sum_profit_loss}")
print(f"Average Change: ${average_change}")
print(f"Greatest Increase in Revenue: {date[max_inc_month]} $({(max_increase)})")
print(f"Greatest Decrease in Revenue: {date[max_dec_month]} $({(max_decrease)})")

# Set output file name
output_path = 'PyBank_output.txt'

# Open the output path as a file object
with open(output_path, 'w') as output:
    # Write analysis to the output file
    output.write(f"Financial Analysis:\n")
    output.write("----------------------------\n")
    output.write(f"Total Months: {count_months}\n")
    output.write(f"Total: {sum_profit_loss}\n")
    output.write(f"Average Change: ${average_change}\n")
    output.write(f"Greatest Increase in Revenue: {date[max_inc_month]} $({(max_increase)})\n")
    output.write(f"Greatest Decrease in Revenue: {date[max_dec_month]} $({(max_decrease)})")
```
